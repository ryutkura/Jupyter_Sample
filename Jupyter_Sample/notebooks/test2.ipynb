````python
import numpy as np
import pandas as pd
````

````python
fd = pd.read_csv("resultdata.csv")
fd
````

````python
maxlen = fd.maxScore
maxlen
````

````python
np.mean(maxlen)
````

````python
import matplotlib.pyplot as plt

# データの確認
print(fd.head())

# 折れ線グラフを描画
plt.figure(figsize=(10, 6))
for column in fd.columns:
    plt.plot(fd[column], label=column)

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot of CSV Data')
plt.legend()
plt.show()
````

````python
# 横の大きさを変数にする
width = len(fd) / 10000  # 例として行数を10000で割った値を横の大きさに設定

# 折れ線グラフを描画
plt.figure(figsize=(width, 6))
for column in fd.columns:
    plt.plot(fd[column], label=column)

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot of CSV Data')
plt.legend()
plt.show()
````

````python
# 横の大きさを変数にする
width = len(fd) / 10000  # 例として行数を10000で割った値を横の大きさに設定

# 折れ線グラフを描画
plt.figure(figsize=(width, 6))
for column in fd.columns:
    plt.plot(fd[column], label=column)

# 横軸の範囲を設定
plt.xlim(0, 100)

# 横軸の目盛りを設定
plt.xticks(range(0, 101, 10))  # 0から1000まで100刻みで目盛りを設定

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot of CSV Data')
plt.legend()
plt.show()
````
